### Audit Table

```sql
CREATE SCHEMA IF NOT EXISTS rawz;

CREATE TABLE IF NOT EXISTS rawz.audit_log (
    file_name STRING,
    timestamp TIMESTAMP,
    load_key STRING,
    completion_status STRING,
    file_path STRING,
    rows_processed INT
)
USING DELTA;
```
